#### Utilizando o mesmo dataset, usei o método PCA para reduzir para uma dimensão.


```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pylab as pl
from itertools import cycle

iris = load_iris() # Carrega o dataset iris, que é um dataset de classificação de flores, da sklearn, utilizado pelo tutor do curso.

numSamples, numFeatures = iris.data.shape # Pega o número de amostras e de features do dataset iris.
```

```python
X = iris.data
pca = PCA(n_components=1, whiten=True).fit(X) # Cria um objeto PCA com 2 componentes e normaliza os dados.
X_pca = pca.transform(X)
```

```python
print(pca.components_) # Mostra os componentes principais.
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
%matplotlib inline
from pylab import *

# Cria um gráfico scatter unidimensional, eixos iguais, com os dados do PCA a fim de visualizar a distribuição dos dados.
colors = cycle('rgb')
target_ids = range(len(iris.target_names))
pl.figure()
for i, c, label in zip(target_ids, colors, iris.target_names):
    pl.scatter(X_pca[iris.target == i, 0], X_pca[iris.target == i, 0],
        c=c, label=label)
pl.legend()
pl.show()
```

#### É possível notar que, mesmo reduzindo de 4D para 1D, os dados ainda são precisos e confiaveis, sem a perda de informações significativas
