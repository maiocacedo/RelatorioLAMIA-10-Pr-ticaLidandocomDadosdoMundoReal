#### Realizando uma simulação com valores diferentes do professor para testar a desconsideração dos outliers

```python

%matplotlib inline
import numpy as np
# Cria uma distribuição normal centrada em 27000 com desvio padrão de 15000, com 10000 elementos
receita = np.random.normal(27000, 15000, 10000)
receita = np.append(receita, np.random.normal(200000, 500000, 100))

# Cria um histograma com 50 barras
import matplotlib.pyplot as plt
plt.hist(receita, 50)
plt.show()
```

```python
# Calcula a média
def reject_outliers(data):
    u = np.median(data) # Média
    s = np.std(data) # Desvio padrão
    filtrado = [e for e in data if (u - 1.5 * s < e < u + 1.5 * s)] # Filtra os outliers
    return filtrado
# Filtra os outliers
filtrado = reject_outliers(receita)

plt.hist(filtrado, 50)
plt.show()
```
